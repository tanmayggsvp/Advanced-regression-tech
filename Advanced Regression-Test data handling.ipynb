```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# to visualise all the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
test_df=pd.read_csv('test.csv')
test_df
```

```python
## First handle Categorical features haveing nan values i.e, which are missing
features_nan=[feature for feature in test_df.columns if test_df[feature].isnull().sum()>1 and test_df[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(test_df[feature].isnull().mean(),4)))
```

```python
## Replace missing value with a new label
def replace_cat_feature(dataset,features_nan):
    data=dataset.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

test_df=replace_cat_feature(test_df,features_nan)
test_df[features_nan].isnull().sum()               # no. of missing values is returned
```

```python
## check for numerical variables containing missing values
numerical_with_nan=[feature for feature in test_df.columns if test_df[feature].isnull().sum()>1 and test_df[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(test_df[feature].isnull().mean(),4)))
```

```python
test_df.shape
```

```python
# Replace missing numerical var values with a new value
for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=test_df[feature].median()
    
    ## create a new feature to capture nan values
    test_df[feature].fillna(median_value,inplace=True)
test_df['LotFrontage'+'nan']=np.where(test_df['LotFrontage'].isnull(),1,0)
test_df['MasVnrArea'+'nan']=np.where(test_df['MasVnrArea'].isnull(),1,0)
test_df['GarageYrBlt'+'nan']=np.where(test_df['GarageYrBlt'].isnull(),1,0)
    
test_df[numerical_with_nan].isnull().sum()
```

### Handling temporal variables

```python
# Date time variables

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       test_df[feature]=test_df['YrSold']-test_df[feature]
```

### Numerical variables
When data is plotted, a few features look like a skewed normal distribution.

```python
# therefore I am performing log normal distribution
import numpy as np
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']

for feature in num_features:
    test_df[feature]=np.log(test_df[feature])
```
