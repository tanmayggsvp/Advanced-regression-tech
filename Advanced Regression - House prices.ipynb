# Feature Engineering technique 

###  To improve the accuracy of prediction, I had used the idea of feature engineering to create a dataset from the given data. In the process, I took care of the following:
    1. Missing values or labels
    2. Temporal variables
    3. Remove rare variables (categorical variables)
    4. Variable values are standardised to the same range.

### Handling missing values

```python
## First handle Categorical features haveing nan values i.e, which are missing
features_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
## Replace missing value with a new label
def replace_cat_feature(dataset,features_nan):
    data=dataset.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

dataset=replace_cat_feature(dataset,features_nan)
dataset[features_nan].isnull().sum()               # no. of missing values is returned
```

```python
## check for numerical variables containing missing values
numerical_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(dataset[feature].isnull().mean(),4)))
```

```python
# Replace missing numerical var values with a new value
for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=dataset[feature].median()
    
    ## create a new feature to capture nan values
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
    
dataset[numerical_with_nan].isnull().sum()
```

```python
dataset.head(25)
```

### Handling temporal variables

```python
# Date time variables

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       dataset[feature]=dataset['YrSold']-dataset[feature]
```

```python
dataset.head()
```

### Numerical variables
When data is plotted, a few features look like a skewed normal distribution.

```python
# therefore I am performing log normal distribution
import numpy as np
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    dataset[feature]=np.log(dataset[feature])
```

```python
final_df=pd.concat([dataset,test_df],axis=0)
final_df.shape
```

### Handling Rare categorical features

```python
categorical_features=[feature for feature in final_df.columns if final_df[feature].dtype=='O']
categorical_features
```

```python
# Removing categorical variables that are present less than 1% of the observations
for feature in categorical_features:
    temp=final_df.groupby(feature)['SalePrice'].count()/len(final_df)
    temp_df=temp[temp>0.01].index
    final_df[feature]=np.where(final_df[feature].isin(temp_df),final_df[feature],'Rare_var')
```

```python
final_df.head(50)
```

```python
# replacing category labels with numbers reflecting category's impact on the sale price

for feature in categorical_features:
    labels_ordered=final_df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    final_df[feature]=final_df[feature].map(labels_ordered)
```

```python
final_df.tail()
```

### Feature Scaling

```python
# Since the range of dataset values are highly different from each other, scaing is done
feature_scale=[feature for feature in final_df.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler       # scikit-learn library
scaler=MinMaxScaler()
scaler.fit(final_df[feature_scale])
```

```python
scaler.transform(final_df[feature_scale])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([final_df[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(final_df[feature_scale]), columns=feature_scale)],
                    axis=1)
```

```python
data.tail()
```

## Feature Selection technique

```python
## for feature selection

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

# to visualise all the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
## Capture the dependent feature
y_final=data[['SalePrice']]
y_train=y_final.iloc[:1460,-1]
```

```python
## drop dependent feature from dataset
X_final=data.drop(['Id','SalePrice'],axis=1)
```

```python
X_train=X_final.iloc[:1460,:]
X_test=X_final.iloc[1460:,:]
```

```python
# First, in the Lasso Regression model, a suitable alpha is selected.
# Bigger the alpha, lesser the values selected
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))  
feature_sel_model.fit(X_train, y_train)
```

```python
feature_sel_model.get_support()      # picks only imp features i.e, assigns non zero coefficients
```

```python
# print the number of total and selected features

selected_feat = X_train.columns[(feature_sel_model.get_support())]

print('total features: {}'.format((X_train.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrank to zero: {}'.format(np.sum(feature_sel_model.estimator_.coef_ == 0)))
```

```python
X_train=X_train[selected_feat]
X_test=X_test[selected_feat]
```
